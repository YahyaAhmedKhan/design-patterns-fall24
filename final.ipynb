````python
# Function to extract wickets from BBM
def extract_wickets_bbm(bbm_value):
    try:
        return int(bbm_value.split('/')[0])
    except:
        return 0

# Function to calculate span in years
def calculate_span_years(span_value):
    try:
        start_year, end_year = map(int, span_value.split('-'))
        return end_year - start_year
    except:
        return 0

# Apply the functions to add derived columns
df['Wickets_BBM'] = df['BBM'].apply(extract_wickets_bbm)
df['Span_Years'] = df['Span'].apply(calculate_span_years)

# Filter records based on the conditions
filtered_df = df[
    (df['Wickets_BBM'] > 11) &
    (df['Mat'] > 15) &
    (df['Span_Years'] > 3)
]

# Display the filtered records
# print(filtered_df)
filtered_df
````

````python
import pandas as pd

def update_player_data(old_df, new_df):
    """
    Removes matching rows from old_df and adds new rows from new_df based on Player names.
    
    Parameters:
    old_df (pandas.DataFrame): Original DataFrame containing all player data
    new_df (pandas.DataFrame): DataFrame containing updated data for some players
    
    Returns:
    pandas.DataFrame: Updated DataFrame with old rows removed and new rows added
    """
    # Create a copy to avoid modifying the original DataFrame
    updated_df = old_df.copy()
    
    # Remove rows where Player names match with new_df
    updated_df = updated_df[~updated_df['Player'].isin(new_df['Player'])]
    
    # Concatenate the remaining old rows with the new rows
    final_df = pd.concat([updated_df, new_df], ignore_index=True)
    
    return final_df


````
